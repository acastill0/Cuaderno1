En base a la simulacion, generar una simulacion lo mas apegado a la realidad del Covid-19, para ello obtener informacion del numero de consultorios y camas disponibles en un hospital del Ecuador. Parametrizar la simulacion para ingresar el numero de pacientes, el numero de consultorios y el numero de camas y algun otro parametro que considere oportuno.

Finalmente generar las siguientes metricas :
- Tiempo de espera promedio para cada sala.
- Tiempo promedio total que requiere un paciente para ser atendido desde cero.
- Cuántos pacientes se atendieron.
- Cuántos pacientes se quedaron sin atender y fallecieron.

```python
import simpy
import random
import matplotlib.pyplot as pp
import numpy as np

%matplotlib inline

#PARAMETROS
HOSPITALES = 1
CAMAS = 115
CONSULTORIOS = 64

INFECTADOS = 229
TASA_CRECIMIENTO = 31
TIEMPO_CONSULTA = 0.015 
DIAS_INTERNADO = 10
DIAS_SIMULACION = 31

persona_recuperadas={}
persona_fallecidas={}
timeEspera=[]
timeConsultas=[]
atendidos=[]
noAtendidosMuertos=[]




class Hospital(object):

    def __init__(self, env, num_cama,num_consultorios, name):
            self.env = env
            self.num_cama = num_cama
            self.num_consultorios=num_consultorios
            self.camas = simpy.Resource(env, num_cama)
            self.consultorios=simpy.Resource(env, num_consultorios)
            self.name = name
            
    def consulta(self,paciente):
        duracion=random.uniform(TIEMPO_CONSULTA-0.005, TIEMPO_CONSULTA+0.005)
        yield self.env.timeout(duracion)
        print('\nPaciente Nº', paciente,'ingreso al cosultorio en tiempo de: ', int(duracion*1440),'min')
        timeConsultas.append(duracion)
        
    def ingresar_paciente(self, paciente):
        yield self.env.timeout(random.randint(DIAS_INTERNADO-5, DIAS_INTERNADO+5))
        print("\nPaciente Nº", paciente, " se interno el día ", int(self.env.now))

def llegada_paciente(env, hospital, paciente):
    arrive = env.now
    with hospital.consultorios.request() as consultorio:
        espera_consulta=random.randint(1,2)
        timeEspera.append(espera_consulta)
        demanda = yield consultorio | env.timeout(espera_consulta)
        llegadaC = env.now - arrive
        if consultorio in demanda:
            print("Paciente Nº", paciente, " se le asigna una cama ")
            yield env.process(hospital.consulta(paciente))
            estado = random.randint(1,100)
            atendidos.append(1)
            if(estado < 40):
                with hospital.camas.request() as cama:
                    llegadaC = env.now
                    dias_esperando = random.randint(1,5) #Numero de dias maximo que puede esperar sin ser asignado a una cama dentro del hospital
                    requerimiento = yield cama | env.timeout(dias_esperando) #Asignar un tiempo de espera
                    wait = env.now - llegadaC
                    if cama in requerimiento:
                        print("Paciente Nº", paciente, " se le asigna una cama ")
                        yield env.process(hospital.ingresar_paciente(paciente))
                        estado = random.randint(1,100)
                        if (estado < 8) :
                            persona_fallecidas[env.now] = persona_fallecidas[env.now] + 1 if env.now in persona_fallecidas else  1
                        else:
                            persona_recuperadas[env.now] = persona_recuperadas[env.now] + 1 if env.now in persona_recuperadas else 1
                    else:
                        print("Paciente Nº" , paciente," espero ", wait , " días y fallece por falta de camas")
                        persona_fallecidas[env.now] = persona_fallecidas[env.now] + 1 if env.now in persona_fallecidas else  1
            else :
                print('Paciente Nº',paciente,'no tiene COVID-19')
        else:
            print('\nPaciente Nº' , paciente,' espero ', llegadaC , " días y fallece por no ingresar a consulta")
            noAtendidosMuertos.append(1)
            persona_fallecidas[env.now] = persona_fallecidas[env.now] + 1 if env.now in persona_fallecidas else  1

def ejecutar(env, tasa_crecimiento, infectados):
    hospital = Hospital(env, CAMAS,CONSULTORIOS, "")
    for i in range(infectados):
        asignar_hospital(env, hospital, i)
    paciente = infectados
    while True:
        yield env.timeout(1)
        for i in range(tasa_crecimiento):
            paciente += 1
            asignar_hospital(env, hospital, paciente)
                
def asignar_hospital(env, hospital, paciente):
    hosp_esc = 1
    if (hosp_esc == 1):
        print("Llega paciente Nº: ", paciente, " en tiempo ", env.now)
        env.process(llegada_paciente(env, hospital, paciente))

print("\t\t\t\t\tSimulacion COVID 19")
env=simpy.Environment()
env.process(ejecutar(env,TASA_CRECIMIENTO, INFECTADOS))
env.run(until=DIAS_SIMULACION)

print("Resultados pacientes :")
print("Recuperados: ")
print(persona_recuperadas)
print("Fallecidos: ")
print(persona_fallecidas)

datos=sorted(persona_recuperadas.items())
x, y =zip(*datos)
pp.plot(x,y,linewidth=2,color='orange') 
pp.scatter(x,y,color='c')
pp.title("Dias  / Personas Recuperadas")
pp.grid(True)
pp.show() 

if (persona_fallecidas):
    datos=sorted(persona_fallecidas.items()) 
    x, y =zip(*datos) 
    pp.plot(x,y,linewidth=2,color='orange') 
    pp.scatter(x,y,color='c')
    pp.title("Personas Fallecidas  / dias de hospitalizacion")
    pp.grid(True) 
    pp.show() 
```

```python
total_atendidos=sum(atendidos)
total_muertos=sum(noAtendidosMuertos)
tiempo_paciente=int(24*(sum(timeEspera)/len(timeEspera)))
tiempo_sala=int(sum(timeConsultas)/len(timeConsultas)*1440)

print("-> Tiempo espera cada sala:",tiempo_sala, 'min')
print("-> Tiempo promedio paciente:",tiempo_paciente,'h')
print("-> Pacientes atendidos:",total_atendidos)
print("-> Pacientes sin atender y fallecieron:",total_muertos)
```
