### Generar los números randómicos

### Ejemplo 1

```python
import pandas as pd, math

iteraciones = int(input ("Iteraciones: "))
semilla = int(input ("Semilla: "))
a=3
b=5
m=19
val_xn=[semilla]
val_un=['']

for i in range(iteraciones): val_xn.append((a*val_xn[-1]+b)%m) ; val_un.append(val_xn[-1]/m)
        
tabla=pd.DataFrame({"XN ":val_xn, "UN ":val_un})
tabla
```
